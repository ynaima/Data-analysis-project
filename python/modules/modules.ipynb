```python
from statistics import mean, mode , median
salaries = [95000, 120000, 105000, 90000, 130000]
mean= mean(salaries)
print(f"mean: {mean}")

mode= mode(salaries)
print(f"mode: {mode}")

median= median(salaries)
print(f"median: {median}")
```
