🟩 Average Salary (1.16.2)

```python
# Code Needed
salaries = [95000, 120000, 105000, 90000, 130000]


def avgSalary(salaries):
    total_salaries= 0
    for salary in salaries:
        total_salaries += salary
    average_salary = total_salaries/(len(salaries))
    return average_salary

avgSalary(salaries)
    
```

🟨 Salary Statistics (1.16.3)

```python
# Code Needed
#mean, medium, min salary, max_salary
salaries = [95000, 120000, 105000, 90000,130000]

def salaryStatistics(salaries):
    mean=0
    medium=0
    min_salary=0
    max_salary=0
    salaries.sort()
    num_salaries=len(salaries)
    
    #finding medium
    if num_salaries % 2 == 0:
            medium = (salaries[(num_salaries // 2)] + salaries[(num_salaries // 2)-1])/2
    else:
            medium= salaries[num_salaries // 2]
        
    #finding mean salary
    mean=sum(salaries)/num_salaries
    
    #finding minimum salary
    min_salary = min(salaries)

    #finding maximum salary
    max_salary = max(salaries)

    return print(f"The salary statistics for the following data is as follows:\n Mean: {mean}\n Medium: {medium}\n Minimum Salary: {min_salary}\n Maximum Salary: {max_salary}")
        


salaryStatistics(salaries) 
```

🟥 Job Posting Summary (1.16.4)

```python

# Code Needed
job_postings = [
    {'title': 'Data Scientist', 'location': 'New York', 'salary': 95000},
    {'title': 'Data Analyst', 'location': 'San Francisco', 'salary': 85000},
    {'title': 'Machine Learning Engineer', 'location': 'New York', 'salary': 115000}
]

def job_summary(job_postings):
    print("Job\t Location\t Salary\t")
    for job in job_postings:
        print(f"{job['title']} {job['location']} {job['salary']}")
    return

job_summary(job_postings)
```
